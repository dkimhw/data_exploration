```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
from pandas.plotting import scatter_matrix

%matplotlib inline
```

### Import & Check Data

```python
recent_grads = pd.read_csv("recent-grads.csv")
```

```python
# Check first row of data
print(recent_grads.iloc[0])
```

```python
# Check first five rows
print(recent_grads.head())
```

```python
# Check nulls
recent_grads.isnull().sum()
```

```python
# Drop nulls (only one missing)
recent_grads = recent_grads.dropna()
```

### EDA

```python
style.use('fivethirtyeight')
```

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", figsize=(8,6))
```

```python
recent_grads.plot(x="Sample_size", y="Unemployment_rate", kind="scatter", figsize=(8,6))
```

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", figsize=(8,6))
```

```python
recent_grads.plot(x="Women", y="Median", kind="scatter", figsize=(8,6))
```

```python
# Histograms (freq)
cols = ["Sample_size", "Median", "Employed", "Full_time"
        , "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(7,14))

for i in range(1, 5):
    ax = fig.add_subplot(4, 1, i)
    ax = recent_grads[cols[i]].plot(kind='hist', rot=40)
    ax.set_title(cols[i])    
    
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(7,14))
for i in range(5,8):
    ax = fig.add_subplot(4,1,i-3)
    ax = recent_grads[cols[i]].plot(kind='hist', rot=40)
    ax.set_title(cols[i])    
fig.tight_layout()
```

```python
# scatter matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
# scatter matrix
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(12,12))
```

```python
# Bar plots on major
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```
